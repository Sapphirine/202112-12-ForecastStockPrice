```python
import os

data_path = 'gdrive/MyDrive/Columbia/BDA_PJ/dataset'
folder_names = list(sorted(os.listdir(data_path)))
print(len(folder_names))
print(folder_names[0])
```

```python
import pandas as pd

def extract_data_from_path(folder_name, data_path):
  stock_index = folder_name.split('_')[0] # stock index
  stock_fnames = os.listdir('{}/{}'.format(data_path, folder_name)) # names of all file under current stock data folder
  stock_tweet_fname = [x for x in stock_fnames if 'dashboard' in x][0] # get dashboard xsl file name

  stock_df = pd.read_excel(open('{}/{}/{}'.format(data_path, folder_name, stock_tweet_fname), 'rb'),
                           sheet_name='Stream')
  stock_df = stock_df.loc[ stock_df['Tweet language (ISO 639-1)']=='en', ['Date', 'Hour', 'Tweet content', 'Followers', 'Hashtags', 'Symbols'] ]
  return stock_index, stock_df
```

```python
dst = 'gdrive/MyDrive/Columbia/BDA_PJ/data'
for idx, folder_name in enumerate(folder_names):
  try:
    stock_index, stock_df = extract_data_from_path(folder_name, data_path)
    stock_df.to_csv('{}/{}.csv'.format(dst, stock_index))
    print('Processed ({}) ${}'.format(idx, stock_index))
  except:
    print('Error ({}) ({})'.format(idx, folder_name))
```
