```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sample = pd.read_csv('data/aal.csv')
sample.head(10)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import string
```

```python
sentence = sample["Tweet content"][4]
print(sentence)
analyzer = SentimentIntensityAnalyzer()
analyzer.polarity_scores(sentence)
```

```python
keys = ['neg', 'neu', 'pos', 'compound']
for k in keys:
    sample[k] = sample['Tweet content'].apply(lambda x : analyzer.polarity_scores(x)[k])
colnames = ['Date', 'Hour', 'Followers', 'neg', 'neu', 'pos', 'compound']
sample = sample[colnames]
```

```python
sample.head()
```

```python
import os
for f in os.listdir('data'):
    df = pd.read_csv(os.path.join('data', f))
    for k in keys:
        df[k] = df['Tweet content'].apply(lambda x : analyzer.polarity_scores(x)[k])
    df = df[colnames]
    df.to_csv(os.path.join('sentiments', f))
```
